```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# we read in the dataset called 'faithful.csv', a record of past eruptions from Yellowstone National Park, USA
df = pd.read_csv('faithful.csv')
df # the dataare the eruption time (eruption ) and the waiting time u
```

```python
plt.scatter(df['eruption'], df['waiting'])
plt.title('Faithful Eruption Dataset')
plt.xlabel('Eruption Length (minutes)')
plt.ylabel('Waiting Time b/w Eruptions (minutes)')
```

```python
X= df.values
X
```

```python
#Unsupervised Algorithm
kmeans=KMeans(n_clusters=2, init='random')
kmeans.fit(X)
```

```python
#the result of our fit
output=kmeans.labels_
output
```

```python
Category0=df[output ==0]
Category1 = df[output==1]

plt.scatter(Category0['eruption'],Category0['waiting'],label = 'Category 0')
plt.scatter(Category1['eruption'], Category1['waiting'], label = 'Category 1')
plt.scatter(3, 75.2, label = 'Category 3') # new Data

plt.title('Faithful Eruption Dataset')
plt.xlabel('Eruption Length (minutes)')
plt.ylabel('Waiting Time b/w Eruptions (minutes)')
```

```python
# Now, imagine we see a new eruption! This one lasts for 3 minutes and came 75.2 minutes after the previous eruption. What category is it in?

kmeans.predict([[3,75.2]])
```

# BULLS EYE DATASET

```python
df1 = pd.read_csv('bulls_eye.csv')[['x', 'y']]
df1
```

```python
plt.scatter(df1['x'], df1['y'])
plt.title('Bulls Eye')
plt.xlabel('x')
plt.ylabel('y')
```

```python
kmeans = KMeans(n_clusters=3, # we have three target regions
                init = 'random',
                max_iter = 300,
                )
kmeans.fit(df1.values)
flags = kmeans.labels_
category0 = df1[flags == 0] # creating a mask that filters for only category 0
category1 = df1[flags == 1] # creating a mask that filters for only category 1
category2 = df1[flags == 2] # creating a mask that filters for only category 2

# plotting the categories
plt.scatter(category0['x'], category0['y'], label = 'Category 0')
plt.scatter(category1['x'], category1['y'], label = 'Category 1')
plt.scatter(category2['x'], category2['y'], label = 'Category 2')

plt.title('Bulls Eye')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc = 'lower right')
```

### that's not what we want! can you think of a way to transform our data to take advantage of the k-means clustering algorithm?

```python
# Polar Coordinates Transformation
#radius
df1['r'] = (df1['x']**2 + df1['y']**2)**0.5 # r^2 = x^2 + y^2
#angle
df1['phi'] = np.arcsin(df1['y'] / df1['r']) # phi = arcsin(y/r)
```

```python
plt.scatter(df1['r'], df1['phi'])
plt.title('Bulls Eye (Polar Coordinates)')
plt.xlabel('r')
plt.ylabel('phi')
```

```python
kmeans = KMeans(n_clusters=3, # we have three target regions
                init = 'random',
                max_iter = 300,
                )
kmeans.fit(df1[['r', 'phi']].values)
flags = kmeans.labels_
category0 = df1[flags == 0] # creating a mask that filters for only category 0
category1 = df1[flags == 1] # creating a mask that filters for only category 1
category2 = df1[flags == 2] # creating a mask that filters for only category 2

# plotting the categories
plt.scatter(category0['x'], category0['y'], label = 'Category 0')
plt.scatter(category1['x'], category1['y'], label = 'Category 1')
plt.scatter(category2['x'], category2['y'], label = 'Category 2')

plt.title('Bulls Eye')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc = 'lower right')
```

```python
len(df1)

 #category 0 - 15 points
 # category 1 = 10 points
 #category 2 - 5 points
```

```python
score =0

for label in kmeans.labels_:
  if label == 0:
    score +=5
  elif label ==1:
    score += 10
  elif label ==2:
    score += 15

print(score)
```
